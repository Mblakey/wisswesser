```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

def create_scatter_plot(dataframe, x_column, y_column):
    sns.scatterplot(data=dataframe, x=x_column, y=y_column)
    plt.show()
```

```python
data = pd.read_csv("~/Workspace/freycompchem/wisswesser/data/smith_wln.tsv",sep="\\t",names=["wln","smiles"],engine='python')
data.head()
```

```python
wln_len = []
smi_len = []

for Index, row in data.iterrows():
    wln = str(row["wln"])
    smiles = str(row["smiles"])
    
    wln_len.append(len(wln))
    smi_len.append(len(smiles))
    
data["wln_len"] = wln_len
data["smiles_len"] = smi_len
```

```python
create_scatter_plot(data,"wln_len","smiles_len")
```
