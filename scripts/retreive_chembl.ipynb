```python
from chembl_webresource_client.new_client import new_client
import pandas as pd
```

```python
chembl_path = "../data/chembl24.tsv"
chembl = pd.read_csv(chembl_path,sep = '\t',names=["WLN","CHEMBL_ID"])
chembl
```

```python
def get_chembl_canonical_smiles(chembl_id):
    try:
        molecule = new_client.molecule
        m1 = molecule.filter(chembl_id=f"{chembl_id}").only(['molecule_structures'])
        return m1[0]['molecule_structures']['canonical_smiles']
    except:
        return "NULL"
```

```python
chembl_smiles = []

for Index, row in chembl.iterrows():
    can_smiles = get_chembl_canonical_smiles(row['CHEMBL_ID'])
    chembl_smiles.append(can_smiles)
```

```python
chembl['can_smiles'] = chembl_smiles
```
